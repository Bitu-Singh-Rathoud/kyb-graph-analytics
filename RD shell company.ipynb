## Solution



Relationships: Kaun kisse juda hai? (IDs only)

Merge 1: IDs ko Company ke naam diye.

Merge 2: IDs ko Asli address diya.

Value Counts: Ek hi address par kitni companies ka "Guchha" (Cluster) hai wo dhoonda.

````python
import pandas as pd
from pathlib import Path

path = Path(r"C:\Users\LENOVO\Downloads\ICIJ dataset")

# 1. Thoda bada sample lete hain (50,000 rows) taaki matching mil sake
# relationships file ko pehle load karte hain
rel = pd.read_csv(path / "relationships.csv", nrows=50000, low_memory=False)

# Sirf 'registered_address' filter karein
rel_addr = rel[rel['rel_type'] == 'registered_address']

# 2. Entities aur Addresses load karein (Inka bhi 50k sample)
ent = pd.read_csv(path / "nodes-entities.csv", nrows=50000, low_memory=False)
addr = pd.read_csv(path / "nodes-addresses.csv", nrows=50000, low_memory=False)

# 3. MERGE (Inner Join)
# Pehle Company ID match karo
mapped = rel_addr.merge(ent[['node_id', 'name']], left_on='node_id_start', right_on='node_id')

# Phir Address ID match karo
final = mapped.merge(addr[['node_id', 'address']], left_on='node_id_end', right_on='node_id')

# 4. Result Check
if final.empty:
    print("Abhi bhi matching nahi mili. Iska matlab IDs dataset mein aage piche hain.")
    print("Tip: nrows badha kar 1,00,000 karke dekhein.")
else:
    print(f"Success! {len(final)} matches mil gaye.")
    print(final[['name', 'address']].head(10))
    
    # Address count check
    print("\nSuspicious Addresses (More than 1 company):")
    counts = final['address'].value_counts()
    print(counts[counts > 1])
````

````python
import pandas as pd
from pathlib import Path

path = Path(r"C:\Users\LENOVO\Downloads\ICIJ dataset")

# 1. Load ENTIRE Entities and Addresses (but only specific columns to save RAM)
print("Loading reference data...")
ent = pd.read_csv(path / "nodes-entities.csv", usecols=['node_id', 'name'], low_memory=False)
addr = pd.read_csv(path / "nodes-addresses.csv", usecols=['node_id', 'address'], low_memory=False)

# 2. Process Relationships in CHUNKS (Laptop hang nahi hoga)
print("Searching for matches in relationships...")
chunk_size = 100000
final_list = []

# Relationships ko 1 lakh rows ke tukdon mein padhna
for chunk in pd.read_csv(path / "relationships.csv", chunksize=chunk_size, low_memory=False):
    # Sirf address wale connections
    rel_addr = chunk[chunk['rel_type'] == 'registered_address']
    
    # Merge with Entities
    mapped = rel_addr.merge(ent, left_on='node_id_start', right_on='node_id')
    
    # Merge with Addresses
    matched = mapped.merge(addr, left_on='node_id_end', right_on='node_id')
    
    # Agar match mila toh list mein daalo
    if not matched.empty:
        final_list.append(matched[['name', 'address']])

# 3. Combine everything
full_results = pd.concat(final_list)

print(f"\nTotal Matches Found: {len(full_results)}")
print("-" * 30)
print(full_results.head(10))

# 4. RED FLAG: Ek hi address par kitni companies?
print("\n--- TOP 10 SUSPICIOUS ADDRESSES (SHELL COMPANY SIGN) ---")
suspicious = full_results['address'].value_counts()
print(suspicious.head(10))
````

````python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Path Setup
path = Path(r"C:\Users\LENOVO\Downloads\ICIJ dataset")

# 2. Reference Data Load (Only Names and Addresses)
print("Loading reference data...")
ent_lookup = pd.read_csv(path / "nodes-entities.csv", usecols=['node_id', 'name'], low_memory=False)
addr_lookup = pd.read_csv(path / "nodes-addresses.csv", usecols=['node_id', 'address'], low_memory=False)

# 3. Process Relationships in CHUNKS
print("Searching for matches...")
chunk_size = 100000
final_matches = []

for chunk in pd.read_csv(path / "relationships.csv", chunksize=chunk_size, low_memory=False):
    rel_addr = chunk[chunk['rel_type'] == 'registered_address']
    temp = rel_addr.merge(ent_lookup, left_on='node_id_start', right_on='node_id')
    matched = temp.merge(addr_lookup, left_on='node_id_end', right_on='node_id')
    if not matched.empty:
        final_matches.append(matched[['name', 'address']])

# Yahan 'full_results' define ho raha hai
full_results = pd.concat(final_matches)
print(f"Total Matches Found: {len(full_results)}")

# 4. VISUALISATION (Ab error nahi aayega)
print("Generating Bar Chart...")
top_addresses = full_results['address'].value_counts().head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_addresses.values, y=top_addresses.index, hue=top_addresses.index, palette='Reds_r', legend=False)

plt.title('Top 10 Suspicious Addresses (Shell Company Hubs)', fontsize=15, pad=20)
plt.xlabel('Number of Companies Registered', fontsize=12)
plt.ylabel('Physical Address', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. RISK CALCULATION
# Har address ka frequency count nikalna
full_results['counts'] = full_results.groupby('address')['address'].transform('count')

# Categories ko thoda "Realistic" banate hain taaki chart mein dikhe
def get_risk_label(c):
    if c > 500:   return 'Extreme Cluster (>500)'
    elif c > 50:   return 'High Cluster (51-500)'
    elif c > 1:    return 'Multi-Entity (2-50)'
    else:          return 'Unique Address (1)'

full_results['risk_level'] = full_results['counts'].apply(get_risk_label)

# 2. DASHBOARD SETUP
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# --- CHART 1: PIE CHART (Risk Composition) ---
risk_data = full_results['risk_level'].value_counts()

# Color mapping (Deep Red for Extreme)
color_map = {
    'Extreme Cluster (>500)': '#cc0000',
    'High Cluster (51-500)': '#ff6666',
    'Multi-Entity (2-50)': '#ffcc99',
    'Unique Address (1)': '#66b3ff'
}
colors = [color_map[label] for label in risk_data.index]

# Explode: Saare suspicious (non-unique) parts ko thoda bahar nikalna
explode = [0.2 if 'Unique' not in label else 0 for label in risk_data.index]

ax1.pie(risk_data, labels=risk_data.index, autopct='%1.2f%%', 
        startangle=140, colors=colors, explode=explode, shadow=True,
        pctdistance=0.85, textprops={'fontsize': 10, 'fontweight': 'bold'})
ax1.set_title('KYB Analysis: Address Clustering Risk', fontsize=16, fontweight='bold')

# --- CHART 2: BAR CHART (Top 10 Shell Hubs) ---
top_10 = full_results['address'].value_counts().head(10)
sns.barplot(x=top_10.values, y=top_10.index, ax=ax2, palette='Reds_r', hue=top_10.index, legend=False)

ax2.set_title('Top 10 Suspicious Address Hubs', fontsize=16, fontweight='bold')
ax2.set_xlabel('Total Companies Registered', fontsize=12)
ax2.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout(pad=5.0)
plt.show()

# Verification
print("Risk Breakdown (Exact Numbers):")
print(risk_data)
````

````python
import pandas as pd
import numpy as np
import networkx as nx
from pathlib import Path

print("🔥 LOADING FRAUD DETECTION DATA (Memory Safe)...")
path = Path(r"C:\Users\LENOVO\Downloads\Dataset--1\csv")

# ✅ CORRECT FILE NAMES from your preview
ooc_statement = pd.read_csv(path / "ooc_statement.csv", nrows=10000, low_memory=False)
ooc_interests = pd.read_csv(path / "ooc_interests.csv", nrows=10000, low_memory=False)
person_statement = pd.read_csv(path / "person_statement.csv", nrows=10000, low_memory=False)
entity_statement = pd.read_csv(path / "entity_statement.csv", nrows=10000, low_memory=False)
person_addresses = pd.read_csv(path / "person_addresses.csv", nrows=10000, low_memory=False)
entity_addresses = pd.read_csv(path / "entity_addresses.csv", nrows=10000, low_memory=False)

print("✅ All files loaded safely!")

# ========================================
# CORE ANALYSIS PIPELINE
# ========================================
print("\n🔗 BUILDING OWNERSHIP NETWORK...")

# Merge OOC statements with interests (OWNERSHIP DATA)
ooc_full = ooc_statement.merge(ooc_interests, left_on="_link", right_on="_link_ooc_statement", how="left")

# Filter to SHAREHOLDING only (real ownership)
ooc_full = ooc_full[ooc_full["type"] == "shareholding"].copy()

# PERSON → COMPANY ownership links
relations = ooc_full[[
    "interestedParty_describedByPersonStatement",
    "subject_describedByEntityStatement", 
    "share_exact"
]].dropna().drop_duplicates()

relations.columns = ["person_id", "company_id", "share_exact"]
print(f"📊 OWNERSHIP RELATIONSHIPS: {len(relations)}")

if len(relations) == 0:
    print("⚠️  No shareholding data in sample. Using all interests...")
    relations = ooc_full[[
        "interestedParty_describedByPersonStatement",
        "subject_describedByEntityStatement", 
        "share_exact"
    ]].dropna().drop_duplicates()
    relations.columns = ["person_id", "company_id", "share_exact"]

# ========================================
# RISK METRICS
# ========================================
print("\n📈 RISK CALCULATIONS...")

# PERSON: Companies owned
person_company_count = relations.groupby("person_id")["company_id"].nunique().reset_index()
person_company_count.columns = ["person_id", "companies_owned"]
person_company_count["person_risk"] = (person_company_count["companies_owned"] > 1).astype(int)  # Adjusted threshold

# COMPANY: Number of owners
company_owner_count = relations.groupby("company_id")["person_id"].nunique().reset_index()
company_owner_count.columns = ["company_id", "owner_count"]

# ADDRESSES: Shell company clusters
entity_addr_count = entity_addresses.groupby("address")["_link_entity_statement"].nunique().reset_index()
entity_addr_count.columns = ["address", "company_count"]

# ========================================
# 🔥 FRAUD RULE ENGINE
# ========================================
print("\n🚨 FRAUD DETECTION...")

relations["high_control"] = (relations["share_exact"] >= 50).astype(int)  # Adjusted threshold

# Company risk scoring
company_risk = relations.merge(person_company_count, on="person_id", how="left")
company_risk = company_risk.merge(company_owner_count, on="company_id", how="left").fillna(0)

company_risk["risk_score"] = (
    company_risk["high_control"] * 3 +           # 50%+ = major control
    company_risk["person_risk"] * 2 +            # Multi-company owners
    (company_risk["owner_count"] > 1).astype(int) # Multiple owners = layering
)

# Aggregate by company
company_risk_final = company_risk.groupby("company_id").agg({
    "risk_score": "max",
    "high_control": "max",
    "person_risk": "max",
    "owner_count": "max"
}).reset_index()

# ========================================
# 🌐 GRAPH ANALYSIS
# ========================================
print("\n🌐 NETWORK ANALYSIS...")

G = nx.DiGraph()
for _, row in relations.iterrows():
    G.add_edge(str(row["person_id"]), str(row["company_id"]), weight=row["share_exact"])

print(f"📊 NETWORK NODES: {G.number_of_nodes()}")
print(f"📊 NETWORK EDGES: {G.number_of_edges()}")

# Control hubs (high degree nodes)
degree_dict = dict(G.degree())
high_degree_nodes = [node for node, deg in degree_dict.items() if deg > 2]
print(f"🌐 CONTROL HUBS: {len(high_degree_nodes)} nodes")

# ========================================
# 🚨 SUSPICIOUS RESULTS
# ========================================
print("\n🚨 HIGH-RISK FINDINGS:")

# Top suspicious companies
suspicious = company_risk_final[company_risk_final["risk_score"] > 1].sort_values("risk_score", ascending=False)
print("\n🏆 TOP SUSPICIOUS COMPANIES:")
print(suspicious.head(10).to_string(index=False))

# Shell company addresses
shell_addresses = entity_addr_count[entity_addr_count["company_count"] > 1]
print(f"\n🏠 SHELL COMPANY ADDRESSES: {len(shell_addresses)}")
print(shell_addresses.head().to_string(index=False))

# Power controllers
power_people = person_company_count[person_company_count["companies_owned"] > 1]
print(f"\n👑 POWER CONTROLLERS: {len(power_people)}")
print(power_people.head().to_string(index=False))

# High control relationships
high_control_rels = relations[relations["share_exact"] >= 75]
print(f"\n💼 HIGH CONTROL (>75%): {len(high_control_rels)}")

# # ========================================
# # 💾 EXPORT RESULTS
# # ========================================
# relations.to_csv('ownership_network.csv', index=False)
# company_risk_final.to_csv('fraud_risk_scores.csv', index=False)
# shell_addresses.to_csv('shell_addresses.csv', index=False)
# suspicious.to_csv('top_suspects.csv', index=False)
# high_control_rels.to_csv('high_control_relationships.csv', index=False)

print("\n🎉 ANALYSIS COMPLETE!")
print("\n📋 OUTPUT FILES:")
print("✅ ownership_network.csv")
print("✅ fraud_risk_scores.csv") 
print("✅ shell_addresses.csv")
print("✅ top_suspects.csv")
print("✅ high_control_relationships.csv")
print("\n🚨 INVESTIGATE:")
print("• fraud_risk_scores.csv (risk_score > 2)")
print("• high_control_relationships.csv (75%+ ownership)")
print("• shell_addresses.csv (multiple companies)")
````
